```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('../../assets/datasets/imdb_p6_sample.csv')
```

```python
bins = [1940, 1950, 1960, 1970, 1980, 1990, 2000, 2010]
group_names = ['40s', '50s', '60s', '70s', '80s', '90s', '00s']

categories = pd.cut(df['year'], bins, labels=group_names)
df['year'] = pd.cut(df['year'], bins, labels=group_names)
```

```python
df_year = pd.get_dummies(df['year'], prefix='Year')
df = df.join(df_year.ix[:, :"Year_90s"])
df.info()
```

```python
nc = [x for x in df.columns if x not in ['rating', 'year', 'tconst', 'title', 'HA']]
X = df[nc]
y = df.rating
```

```python
from sklearn.cross_validation import cross_val_score, StratifiedKFold
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, ExtraTreesClassifier, BaggingClassifier
```

```python
dt = DecisionTreeClassifier(class_weight='balanced')
cv = StratifiedKFold(y, n_folds=3, shuffle=True, random_state=21)
```

```python
def cross_val(model):
    s = cross_val_score(model, X, y, cv=cv, n_jobs=-1)
    print("{} Score:\t{:0.3} Â± {:0.3}".format(type(model), s.mean().round(3), s.std().round(3)))
```

```python
cross_val(dt)
```
