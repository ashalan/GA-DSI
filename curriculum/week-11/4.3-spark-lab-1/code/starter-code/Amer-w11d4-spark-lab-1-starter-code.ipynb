```python
trips = sc.textFile('file:///home/vagrant/data/201408_babs_open_data/201408_trip_data.csv')
```

```python
type(trips)
```

```python
splitTrips = trips.map(lambda line: [x for x in line.split(',')])
```

```python
st = splitTrips.collect()
```

```python
station_70 = splitTrips.filter(lambda line: "70" in line[4])
```

```python
station_70.collect()
```

```python
station_70.cache()
```

```python
by_hour = station_70.map(lambda line: line[2].split()).map(lambda lst: ((lst[0], lst[1].split(':')[0]), 1))
```

```python
by_hour_2 = by_hour.reduceByKey(lambda a, b: a+b)
```

```python
hours = by_hour_2.map(lambda tup: (int(tup[0][1]), float(tup[1])))
```

```python
hours.collect()
```

```python
sumCount = hours.combineByKey(lambda value: (value, 1),
                             lambda x, value: (x[0] + value, x[1] + 1),
                             lambda x, y: (x[0] + y[0], x[1] + y[1]))

averageByKey = sumCount.map(lambda (label, (value_sum, count)): (label, value_sum / count))

averageByKey.collect()
```

```python
averageByKey.sortBy(lambda x: x[1], ascending=False).collect()
```
