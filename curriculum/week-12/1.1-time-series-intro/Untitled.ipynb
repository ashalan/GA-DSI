```python
sp = web.DataReader('^GSPC','yahoo')
```

```python
sp.head()
```

```python
for x in range(1,31):
    sp['close_prior_'+str(x)] = sp['Close'].shift(x)
```

```python
sp.head()
```

```python
sp['Tomorrow'] = sp['Close'].shift(-1)
```

```python
sp = sp.dropna()
```

```python
sp.head(10)
```

```python
from sklearn.svm import SVR
```

```python
y = sp['Tomorrow']
X = sp[[x for x in sp.columns if x != 'Tomorrow']]
clf = SVR(C=1.0, epsilon=0.2)
clf.fit(X, y) 
```
