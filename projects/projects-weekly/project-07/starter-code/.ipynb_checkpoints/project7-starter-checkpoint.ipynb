#### 4.3 What is the value of understanding correlations before PCA? 

A set of correlated variables might load onto several principal components (eigenvectors),so including many variables from such a set will differentially weight several eigenvectors, and thereby change the directions of all eigenvectors, too.  So, according to these considerations, we should discard some highly correlated variables before doing PCA.

### Part 5: Perform a PCA

#### 5.1 Conduct the PCA

```python
covMat = np.cov(xStand.T)
eigenValues2, eigenVectors2 = np.linalg.eig(covMat)
```

```python
eigenPairs2 = [(np.abs(eigenValues2[i]), eigenVectors2[:,i]) for i in range(len(eigenValues2))]
eigenPairs2.sort()
eigenPairs2.reverse()
for i in eigenPairs2:
    print(i[0])
```

```python
totalEigen = sum(eigenValues2)
varExpl = [(i / totalEigen)*100 for i in sorted(eigenValues2, reverse=True)]
cumulvarExpl = np.cumsum(varExpl)
```

```python
print(cumulvarExpl)
```

```python
airports_pca = PCA(n_components=7)
airports_pca.fit(xStand)
X = airports_pca.transform(xStand)
X_pca = airports_pca.fit_transform(xStand)
```

```python
from matplotlib.colors import ListedColormap
def plot_decision_regions(X, y, classifier, test_idx=None, resolution=0.02):
    markers = ('s', 'x', 'o', '^', 'v')
    colors = ('red', 'blue', 'lightgreen', 'gray', 'cyan')
    cmap = ListedColormap(colors[:len(np.unique(y))])
    # plot the decision surface
    x1_min, x1_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    x2_min, x2_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx1, xx2 = np.meshgrid(np.arange(x1_min, x1_max, resolution),
    np.arange(x2_min, x2_max, resolution))
    Z = classifier.predict(np.array([xx1.ravel(), xx2.ravel()]).T)
    Z = Z.reshape(xx1.shape)
    plt.contourf(xx1, xx2, Z, alpha=0.4, cmap=cmap)
    plt.xlim(xx1.min(), xx1.max())
    plt.ylim(xx2.min(), xx2.max())
    # plot all samples
    X_test, y_test = X[test_idx, :], y[test_idx]
    for idx, cl in enumerate(np.unique(y)):
        plt.scatter(x=X[y == cl, 0], y=X[y == cl, 1], alpha=0.8, c=cmap(idx), marker=markers[idx], label=cl)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
```

```python
lr.fit(X, y)
```

```python
plot_decision_regions(X_pca, y, classifier=lr)
```
